```python
import os
import numpy as np
from tensorflow.keras.preprocessing import image
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import matplotlib.pyplot as plt

# Load the saved model
loaded_model = load_model('imageclassifier_10.h5')

# Image dimensions
img_width, img_height = 224, 224  # Set your desired image dimensions

# Get user input for the folder containing images
folder_path = input("Enter the folder path containing the images: ")

# Check if the folder exists
if not os.path.exists(folder_path):
    print(f"Error: The folder {folder_path} does not exist.")
    exit()

# List all files in the folder
image_files = [f for f in os.listdir(folder_path) if os.path.isfile(os.path.join(folder_path, f))]

# Filter out allowed image extensions
allowed_extensions = ['.jpg', '.jpeg', '.png', '.gif', '.bmp', '.tiff']
image_files = [f for f in image_files if any(f.lower().endswith(ext) for ext in allowed_extensions)]

# Check if there are images in the folder
if not image_files:
    print(f"No images found in the folder {folder_path}.")
    exit()

# Iterate through the images
for image_file in image_files:
    image_path = os.path.join(folder_path, image_file)

    # Load the image for prediction
    img = image.load_img(image_path, target_size=(img_width, img_height))
    img = image.img_to_array(img)
    img = np.expand_dims(img, axis=0)
    img = img / 255.0  # Normalize the pixel values

    # Predict the class of the image using the loaded model
    predicted_class = np.argmax(loaded_model.predict(img), axis=-1)
    class_labels = ['Belly Dance', 'BharatNatyam', 'Flamenco', 'Jazz', 'Kathakali', 'Lion Dance', 'Waltz', 'Folklorico', 'Hiphop', 'Samba']
    predicted_class_name = class_labels[predicted_class[0]]

    # Create folder if it doesn't exist
    output_folder = os.path.join(folder_path, predicted_class_name)
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)

    # Move the image to the respective class folder
    new_image_path = os.path.join(output_folder, image_file)
    os.replace(image_path, new_image_path)

    # Display the class of the image
    print(f"Image '{image_file}' belongs to class: {predicted_class_name}")

# Display images with their predicted classes
fig, axs = plt.subplots(4, 5, figsize=(15, 10))
axs = axs.ravel()
for i in range(56):
    if i < len(image_files):
        img = image.load_img(os.path.join(folder_path, image_files[i]))
        axs[i].imshow(img)
        axs[i].set_title(predicted_class_name)
        axs[i].axis('off')
    else:
        axs[i].axis('off')

plt.tight_layout()
plt.show()
```

```python
import os
import numpy as np
from tensorflow.keras.preprocessing import image
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import matplotlib.pyplot as plt

# Load the saved model
loaded_model = load_model('imageclassifier_10.h5')

# Image dimensions
img_width, img_height = 224, 224  # Set your desired image dimensions

# Get user input for the folder containing images
folder_path = input("Enter the folder path containing the images: ")

# Check if the folder exists
if not os.path.exists(folder_path):
    print(f"Error: The folder {folder_path} does not exist.")
    exit()

# Ask user for the name of the parent folder
parent_folder_name = input("Enter the name of the parent folder to be created: ")
parent_folder_path = os.path.join(folder_path, parent_folder_name)

# Create the parent folder
os.makedirs(parent_folder_path, exist_ok=True)

# List all files in the folder
image_files = [f for f in os.listdir(folder_path) if os.path.isfile(os.path.join(folder_path, f))]

# Filter out allowed image extensions
allowed_extensions = ['.jpg', '.jpeg', '.png', '.gif', '.bmp', '.tiff']
image_files = [f for f in image_files if any(f.lower().endswith(ext) for ext in allowed_extensions)]

# Check if there are images in the folder
if not image_files:
    print(f"No images found in the folder {folder_path}.")
    exit()

# Iterate through the images
for image_file in image_files:
    image_path = os.path.join(folder_path, image_file)

    # Load the image for prediction
    img = image.load_img(image_path, target_size=(img_width, img_height))
    img = image.img_to_array(img)
    img = np.expand_dims(img, axis=0)
    img = img / 255.0  # Normalize the pixel values

    # Predict the class of the image using the loaded model
    predicted_class = np.argmax(loaded_model.predict(img), axis=-1)
    class_labels = ['Belly Dance', 'BharatNatyam', 'Flamenco', 'Jazz', 'Kathakali', 'Lion Dance', 'Waltz', 'Folklorico', 'Hiphop', 'Samba']
    predicted_class_name = class_labels[predicted_class[0]]

    # Create folder if it doesn't exist within the parent folder
    output_folder = os.path.join(parent_folder_path, predicted_class_name)
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)

    # Move the image to the respective class folder within the parent folder
    try:
        new_image_path = os.path.join(output_folder, image_file)
        os.replace(image_path, new_image_path)
    except FileNotFoundError:
        print(f"Warning: Image '{image_file}' could not be found.")

    # Display the class of the image
    print(f"Image '{image_file}' belongs to class: {predicted_class_name}")

# Display images with their predicted classes
fig, axs = plt.subplots(4, 5, figsize=(15, 10))
axs = axs.ravel()
for i in range(56):
    if i < len(image_files):
        img = image.load_img(os.path.join(folder_path, image_files[i]))
        axs[i].imshow(img)
        axs[i].set_title(predicted_class_name)
        axs[i].axis('off')
    else:
        axs[i].axis('off')

plt.tight_layout()
plt.show()
```
